```python
import pandas as pd
```

```python
df = pd.read_csv('Pre- Exponential Factor.csv')
```

```python
# for straw without catalyst
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
spline = make_interp_spline(df['α'], df['DAEM'], k=3)
smooth_x = np.linspace(df['α'].min(), df['α'].max(), 300)
smooth_y = spline(smooth_x)
spline = make_interp_spline(df['α'], df['DAEM'], k=3)
plt.scatter(df['α'], df['DAEM'],marker="^")
plt.plot(smooth_x, smooth_y,"--",label ='KAS:Without catalyst')
spline = make_interp_spline(df['α'], df['FWO'], k=3)
smooth_x = np.linspace(df['α'].min(), df['α'].max(), 300)
smooth_y = spline(smooth_x)
spline = make_interp_spline(df['α'], df['FWO'], k=3)
plt.scatter(df['α'], df['FWO'],marker="^")
plt.plot(smooth_x, smooth_y,"--" ,label ='KAS:with NiCaOSi$_2$O$_4$ Catalyst')
spline = make_interp_spline(df['α'], df['Friedman'], k=3)
smooth_x = np.linspace(df['α'].min(), df['α'].max(), 300)
smooth_y = spline(smooth_x)
spline = make_interp_spline(df['α'], df['Friedman'], k=3)
plt.scatter(df['α'], df['Friedman'],marker="*")
plt.plot(smooth_x, smooth_y,"--",label ='KAS:With NiCa$_2$SiO$_4$ Catalyst' )
plt.xticks(fontsize=8, weight='bold')
plt.yticks(fontsize=8, weight='bold')
plt.xlabel('α', fontsize=12, weight='bold')
plt.ylabel('LnK$_o$', fontsize=12, weight='bold')
plt.title('Straw',weight='bold')
plt.legend(prop={'weight': 'bold'})
```
