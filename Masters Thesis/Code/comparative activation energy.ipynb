```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['α'],df['DAEM'])
plt.plot(df['α'],df['FWO'])
plt.plot(df['α'],df['Friedman'])
plt.scatter(df['α'],df['DAEM'])
plt.scatter(df['α'],df['FWO'])
plt.scatter(df['α'],df['Friedman'])
```

```python
# for straw without catalyst
import numpy as np
from scipy.interpolate import make_interp_spline
spline = make_interp_spline(df['α'], df['DAEM'], k=3)
smooth_x = np.linspace(df['α'].min(), df['α'].max(), 300)
smooth_y = spline(smooth_x)
spline = make_interp_spline(df['α'], df['DAEM'], k=3)
plt.scatter(df['α'], df['DAEM'])
plt.plot(smooth_x, smooth_y, label ='DAEM')
spline = make_interp_spline(df['α'], df['FWO'], k=3)
smooth_x = np.linspace(df['α'].min(), df['α'].max(), 300)
smooth_y = spline(smooth_x)
spline = make_interp_spline(df['α'], df['FWO'], k=3)
plt.scatter(df['α'], df['FWO'])
plt.plot(smooth_x, smooth_y, label ='FWO')
spline = make_interp_spline(df['α'], df['Friedman'], k=3)
smooth_x = np.linspace(df['α'].min(), df['α'].max(), 300)
smooth_y = spline(smooth_x)
spline = make_interp_spline(df['α'], df['Friedman'], k=3)
plt.scatter(df['α'], df['Friedman'])
plt.plot(smooth_x, smooth_y, label ='Friedman' )
plt.xticks(fontsize=8, weight='bold')
plt.yticks(fontsize=8, weight='bold')
plt.xlabel('α', fontsize=12, weight='bold')
plt.ylabel('$E_a$', fontsize=12, weight='bold')
plt.title('Straw Without Catalyst')
plt.legend(prop={'weight': 'bold'})
```
